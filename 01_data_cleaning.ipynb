```python
current_year = 2025

df["VEHICLE AGE"] = current_year - df["VEHICLE MAKE YEAR"]

df["PREMIUM RATE"] = df["POLICY PREMIUM"] / df["SUM INSURED"]

df.head()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["POLICY PREMIUM"], bins=50)

plt.title("Distribution of Policy Premium")
plt.xlabel("Premium")
plt.ylabel("Frequency")

plt.show()
```
